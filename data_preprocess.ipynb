```python
import pandas as pd
```

```python
df_standing = pd.read_csv('preprocessed/standing.csv')
df_sitting = pd.read_csv('preprocessed/Sitting.csv')
df_fall1 = pd.read_csv('preprocessed/falldown.csv')
df_fall2 = pd.read_csv('preprocessed/falldown2.csv')
```

```python
df_list = [df_standing, df_sitting, df_fall1, df_fall2]
```

```python
df_standing
```

```python
for df in df_list:
    print(df.shape)
```

```python
df_standing['output'] = 'standing'
df_sitting['output'] = 'sitting'
df_fall1['output'] = 'falling'
df_fall2['output'] = 'falling'
```

```python
df = pd.concat(df_list)
df
```

```python
output_dict = {'standing':0, 'falling':1, 'sitting':2}
df['mapped_output'] = df['output'].map(output_dict)
```

```python
df[['mapped_output', 'output']][1000:]
```

```python
df.drop('output', axis=1, inplace=True)
```

```python
df = df.sample(frac=1)
```

```python
df.to_csv("Final/dataset.csv", index = False)
```
